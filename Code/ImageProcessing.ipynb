```
# Imports
import cv2
from google.colab.patches import cv2_imshow
```

```
# Data Directory
DATA_DIR = '/content/drive/My Drive/capstone_data'
```

```
# Sample image
img_path = DATA_DIR + '/Places365_test_00000003.jpg'
```

```
# Originial image
img = cv2.imread(img_path)
cv2_imshow(img)
```

```
# Convert BGR to LAB color space
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2_imshow(img_lab)
```

```
# Split into L and ab (Note: cv2.split() is expensive operation)

L = img_lab[:, :, 0]
ab = img_lab[:, :, 1:]
```

```
# L channel
cv2_imshow(L)
```

```
# Merge L and ab channels, to recreate Lab color space
merged_img = cv2.merge((L,ab))
cv2_imshow(merged_img)
```

```
# Convert LAB to BGR color space (To get original image back)
img_bgr = cv2.cvtColor(merged_img, cv2.COLOR_LAB2BGR)
cv2_imshow(img_bgr)
```
